```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
archivo = pd.read_excel('proyecto.xlsx')
```

```python

aos = archivo['Ao']
ipcChina = archivo['IPC(%)CHINA']
ipcUsa = archivo['IPC (%)USA']
ipcJAPON = archivo['IPC (%)JPN']
ipcRUSIA = archivo['IPC (%)RUS']
ipcESP = archivo['IPC (%)ESP']
# Graficar los datos
plt.plot(aos, ipcChina, label='IPC (%) China', color ='red')
plt.plot(aos, ipcJAPON, label='IPC (%) Japon', color ='yellow')
plt.plot(aos, ipcRUSIA, label='IPC (%) Rusia', color ='black')
plt.plot(aos, ipcESP, label='IPC (%) Espaa', color ='orange')
plt.plot(aos, ipcUsa, label='IPC (%) Estados Unidos')

# Aadir etiquetas y ttulo
plt.xlabel('Ao')
plt.ylabel('IPC (%)')
plt.title('Comparacin del IPC')
plt.legend()
plt.show()

```
