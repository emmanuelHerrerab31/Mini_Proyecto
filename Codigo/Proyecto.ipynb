```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
archivo = pd.read_excel('proyecto.xlsx')
```

```python

años = archivo['Año']
ipcChina = archivo['IPC(%)CHINA']
ipcUsa = archivo['IPC (%)USA']
ipcJAPON = archivo['IPC (%)JPN']
ipcRUSIA = archivo['IPC (%)RUS']
ipcESP = archivo['IPC (%)ESP']
# Graficar los datos
plt.plot(años, ipcChina, label='IPC (%) China', color ='red')
plt.plot(años, ipcJAPON, label='IPC (%) Japon', color ='yellow')
plt.plot(años, ipcRUSIA, label='IPC (%) Rusia', color ='black')
plt.plot(años, ipcESP, label='IPC (%) España', color ='orange')
plt.plot(años, ipcUsa, label='IPC (%) Estados Unidos')

# Añadir etiquetas y título
plt.xlabel('Año')
plt.ylabel('IPC (%)')
plt.title('Comparación del IPC')
plt.legend()
plt.show()

```
